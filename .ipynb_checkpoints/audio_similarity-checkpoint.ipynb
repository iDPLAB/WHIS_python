```python
import numpy as np
import soundfile as sf

def audio_wave_similarity(path1, path2):
    x1, fs1 = sf.read(path1)
    x2, fs2 = sf.read(path2)

    assert fs1 == fs2
    min_len = min(len(x1), len(x2))
    x1 = x1[:min_len]
    x2 = x2[:min_len]

    mse = np.mean((x1 - x2) ** 2)
    corr = np.corrcoef(x1, x2)[0, 1]

    # print(f"MSE = {mse:.2e}")
    # print(f"Correlation = {corr:.6f}")
    print(f"{mse:.2e}      ", end=' ')
    # print(f"{corr:.6f}        ", end=' ')
    return

```

```python
import librosa
import numpy as np
from scipy.spatial.distance import cosine

def audio_perceptual_similarity(path1, path2):
    x1, fs1 = librosa.load(path1, sr=None)
    x2, fs2 = librosa.load(path2, sr=None)

    assert fs1 == fs2

    mfcc1 = librosa.feature.mfcc(y=x1, sr=fs1, n_mfcc=13)
    mfcc2 = librosa.feature.mfcc(y=x2, sr=fs2, n_mfcc=13)

    v1 = np.mean(mfcc1, axis=1)
    v2 = np.mean(mfcc2, axis=1)

    sim = 1 - cosine(v1, v2)
    # print(f"MFCC similarity = {sim:.4f}")
    print(f"{sim:.6f}      ", end=' ')
    return

    
```

```python
HL = ["HL1", "HL2", "HL3", "HL4", "HL5", "HL6", "HL7", "HL8"]
df = ["dtvf", "fabs"]
cmprs = ["0", "50", "100"]
snd_name = "Snd_Hello123"
path = "Sound/"
for i in HL:
    print("                     MSE     MFCC_similarity")
    for j in cmprs:
        for k in df:
            audio1 = path + snd_name + "_WHISv302" + k + "_" + i + "_Cmprs" + j + ".wav"
            audio2 = path + snd_name + "_" + i + "_Cmprs" + j + "_WHISv302" + k + ".wav"
            print((i + k + "Cmprs" + j +":").ljust(17), end=' ')
            audio_wave_similarity(audio1, audio2)
            audio_perceptual_similarity(audio1, audio2)
            print()

    print()
```
